```python
# Continuous - Age
plt.figure(figsize=(8, 6)) 
plt.hist(df['age'], bins=17, color='lightgreen', edgecolor='black') 
plt.title('Histogram of Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(axis='y') 
plt.show()
```
